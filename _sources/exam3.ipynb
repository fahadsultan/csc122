class Stack:
    def __init__(self):
        self.__stack = []
    
    def push(self, x):
        self.__stack.append(x)
    
    def pop(self):
        return self.__stack.pop()
    
    def top(self):
        return self.__stack[-1]
    
    def empty(self):
        return len(self.__stack) == 0

# **Question 2.** Book Class (25 pts.)

<u>All attributes must be **private**</u> and inaccessible from outside the class directly. In other words, all interactions with the class should be done through methods.

The class should have the following **instance attributes**:

- `title` (str): The title of the book.
- `author` (str): The author of the book.
- `year` (int): The year the book was published.
- `price` (str): The genre of the book.

The class should have the following **class attributes**:

- `book_count` (int): The total number of books created.
- `discount_rate` (float): The discount rate for all books. The default value is 0.1, which means a 10% discount.

The class should have the following **instance methods**:

- `__init__(self, title, author, year, genre)`: The constructor method that initializes the instance attributes.
- `get_price(self)`: A method that returns the price of the book, AFTER applying the discount rate.

The class should have the following **class methods**:

- `get_book_count(cls)`: A class method that returns the total number of books created.

The class should have the following **static methods**:

- `is_valid_year(year)`: A static method that returns `True` if the year is between 1450 and 2024, and `False` otherwise.

```python
class Book: 
    discount_rate = 0.1
    n_books = 0

    def __init__(self, title, author, price):
        self.title = title
        self.author = author
        self.price = price
        Book.n_books += 1

    def get_price(self):
        return self.price - self.price * Book.discount_rate
    
    @classmethod
    def get_book_count(cls):
        return cls.n_books
    
    @staticmethod
    def is_valid_year(year):
        return True if (year > 1440) and (year <= 2024) else False
    
book1 = Book("The Reluctant Fundamentalist", "Mohsin Hamid", 500)
book2 = Book("The Kite Runner", "Khaled Hosseini", 450)
book3 = Book("Program or be Programmed", "Douglas Rushkoff", 300)

assert Book.get_book_count()    == 3, "Test case 1 failed"
assert Book.is_valid_year(2021) == True, "Test case 2 failed"
assert Book.is_valid_year(1440) == False, "Test case 3 failed"
assert book1.get_price()        == 450, "Test case 4 failed"
assert book2.get_price()        == 405, "Test case 5 failed"
assert book3.get_price()        == 270, "Test case 6 failed"

print("All test cases passed successfully.")
```

# **Question 3**: Design Browser History (25 pts.)

You have a browser of one tab where you can perform one of the following operations: 

* Start on the homepage (string) 
* Visit another url (also string)
* Go back in the history a specific number of steps 
* Move forward in the history a specific number of steps.

Implement the BrowserHistory class:

* `__init__(string homepage)` Initializes the object with the homepage (string) of the browser. 

* `visit(self, url)` (Instance method) Adds a new url to the history. Clears up all the forward history. This method does not return anything.

* `back(self, n)` (Instance method) Move `n` (integer) steps back in history. If you can only back `x` steps in the history and `n > x`, you will back only `x` steps. <u>Return the current url</u> after moving back in history at most `n` steps.

* `forward(self, n)` (Instance method) Move `n` steps forward in history. If you can only forward `x` steps in the history and `n > x`, you will forward only `x` steps. <u>Return the current url</u> after moving forward in history

```python
class BrowserHistory: 
    
    def __init__(self, homepage: str):
        self.history = Stack()
        self.future  = Stack()
        self.history.push(homepage)

    def visit(self, url: str) -> None:
        self.history.push(url)
        self.future = Stack()

    def back(self, steps: int) -> str:
        while steps > 0 and not self.history.empty():
            self.future.push(self.history.pop())
            steps -= 1
        return self.history.top()
    
    def forward(self, steps: int) -> str:
        while steps > 0 and not self.future.empty():
            self.history.push(self.future.pop())
            steps -= 1
        return self.history.top()

hist = BrowserHistory("leetcode.com")
hist.visit("google.com")
hist.visit("facebook.com")
hist.visit("youtube.com")

assert hist.back(1) == "facebook.com"
assert hist.back(1) == "google.com"
assert hist.forward(1) == "facebook.com"
assert hist.visit("linkedin.com") == None

print("All tests passed.")
```
